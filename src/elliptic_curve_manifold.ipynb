```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_discriminant_and_roots():
    """
    Calculates and plots the discriminant and roots of the cubic x^3 - x + 1
    """
    # Calculate discriminant of x^3 - x + 1
    # For a cubic ax^3 + bx^2 + cx + d, discriminant is:
    # b^2c^2 - 4ac^3 - 4b^3d - 27a^2d^2 + 18abcd
    # Here a=1, b=0, c=-1, d=1
    # Discriminant = 0 - 4(-1)^3 - 0 - 27(1)(1)^2 + 0 = 4 - 27 = -23

    print("For the cubic x³ - x + 1:")
    print(
        "Discriminant = -23 (indicating one real root and two complex conjugate roots)"
    )

    # Find roots numerically
    coefficients = [1, 0, -1, 1]  # x^3 - x + 1
    roots = np.roots(coefficients)

    print("\nRoots of x³ - x + 1 = 0:")
    for i, root in enumerate(roots):
        print(f"Root {i + 1}: {root:.4f}")

    # Plot the roots in the complex plane
    plt.figure(figsize=(8, 6))
    plt.scatter(
        np.real(roots), np.imag(roots), color="red", s=100, label="Roots of $x^3-x+1=0$"
    )
    plt.axhline(0, color="k", linestyle="--", alpha=0.5)
    plt.axvline(0, color="k", linestyle="--", alpha=0.5)
    plt.xlabel("Real Part")
    plt.ylabel("Imaginary Part")
    plt.title("Roots of $x^3 - x + 1 = 0$ in Complex Plane")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()


print("Visualizing the elliptic curve y² = x³ - x + 1")
print("=" * 50)

# Show information about the curve
plot_discriminant_and_roots()
```

```python
def plot_elliptic_curve_with_real_points():
    """
    Plot the real points of the elliptic curve y^2 = x^3 - x + 1
    """
    # Define x values in the range where x^3 - x + 1 >= 0
    # From our root analysis, we know the real root is approximately -1.3247
    # So we'll plot from that point onwards
    x_real_root = -1.3247
    x = np.linspace(x_real_root, 3, 1000)

    # Calculate y^2 = x^3 - x + 1
    y_squared = x**3 - x + 1

    # Calculate positive and negative y values
    y_positive = np.sqrt(y_squared)
    y_negative = -np.sqrt(y_squared)

    plt.figure(figsize=(10, 8))
    plt.plot(x, y_positive, "b-", label=r"$y = +\sqrt{x^3-x+1}$", linewidth=2)
    plt.plot(x, y_negative, "r-", label=r"$y = -\sqrt{x^3-x+1}$", linewidth=2)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Real Points of Elliptic Curve: $y^2 = x^3 - x + 1$")
    plt.grid(True, alpha=0.3)
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.legend()
    plt.xlim(x_real_root - 0.2, 3)
    plt.show()


print("\nPlotting real points of the curve...")
plot_elliptic_curve_with_real_points()
```

```python
def plot_elliptic_curve_real_part():
    """
    Visualizes the real part of the elliptic curve y² = x³ - x + 1.
    The curve is plotted where the right side of equation is non-negative.
    """
    # Define x range focusing on where x^3 - x + 1 >= 0
    x = np.linspace(-1.5, 3, 400)
    y = np.linspace(-3, 3, 400)
    X, Y = np.meshgrid(x, y)

    # Calculate the right-hand side of the equation: x^3 - x + 1
    Z_squared = X**3 - X + 1

    # Only consider where z^2 >= 0 for real y values
    valid_region = Z_squared >= 0

    # Calculate positive and negative square roots
    Z_pos = np.where(valid_region, np.sqrt(Z_squared), np.nan)
    Z_neg = np.where(valid_region, -np.sqrt(Z_squared), np.nan)

    # Plot
    fig = plt.figure(figsize=(14, 6))

    # Positive part
    ax1: Axes3D = fig.add_subplot(121, projection="3d")
    surface1 = ax1.plot_surface(
        X, Y, Z_pos, alpha=0.7, cmap="viridis", edgecolor="none"
    )
    ax1.set_title("Positive Part of y² = x³ - x + 1")
    ax1.set_xlabel("X")
    ax1.set_ylabel("Y")
    ax1.set_zlabel("Y values")
    fig.colorbar(surface1, ax=ax1, shrink=0.5, aspect=20)

    # Negative part
    ax2: Axes3D = fig.add_subplot(122, projection="3d")
    surface2 = ax2.plot_surface(X, Y, Z_neg, alpha=0.7, cmap="plasma", edgecolor="none")
    ax2.set_title("Negative Part of y² = x³ - x + 1")
    ax2.set_xlabel("X")
    ax2.set_ylabel("Y")
    ax2.set_zlabel("Y values")
    fig.colorbar(surface2, ax=ax2, shrink=0.5, aspect=20)

    plt.tight_layout()
    plt.show()


print("\nPlotting real parts of the curve in 3D...")
plot_elliptic_curve_real_part()
```

```python
def visualize_complex_elliptic_curve():
    """
    Visualization of the elliptic curve y² = x³ - x + 1 in a complex domain context.
    Shows the real part and some properties of the curve.
    """
    # Create a complex grid for visualization purposes
    real_x = np.linspace(-2, 3, 400)
    imag_x = np.linspace(-1.5, 1.5, 400)
    real_X, imag_X = np.meshgrid(real_x, imag_x)

    # Complex x values
    X_complex = real_X + 1j * imag_X

    # Calculate x^3 - x + 1 for complex x
    expr_values = X_complex**3 - X_complex + 1

    # Calculate |y| = |sqrt(x^3 - x + 1)| for magnitude visualization
    y_magnitude = np.abs(expr_values)

    # Calculate arg(y) for phase visualization
    y_phase = np.angle(expr_values)

    # Plot magnitude and phase
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))

    # Plot 1: Magnitude of y^2 = x^3 - x + 1
    im1 = axes[0, 0].contourf(real_X, imag_X, y_magnitude, levels=50, cmap="viridis")
    axes[0, 0].set_title("Magnitude of $x^3 - x + 1$")
    axes[0, 0].set_xlabel("Real Part of x")
    axes[0, 0].set_ylabel("Imaginary Part of x")
    plt.colorbar(im1, ax=axes[0, 0])

    # Plot 2: Phase of y^2 = x^3 - x + 1
    im2 = axes[0, 1].contourf(
        real_X, imag_X, y_phase, levels=50, cmap="twilight", vmin=-np.pi, vmax=np.pi
    )
    axes[0, 1].set_title("Phase of $x^3 - x + 1$")
    axes[0, 1].set_xlabel("Real Part of x")
    axes[0, 1].set_ylabel("Imaginary Part of x")
    plt.colorbar(im2, ax=axes[0, 1])

    # Plot 3: Real part of the expression
    real_expr = np.real(expr_values)
    im3 = axes[1, 0].contourf(real_X, imag_X, real_expr, levels=50, cmap="RdBu")
    axes[1, 0].set_title("Real Part of $x^3 - x + 1$")
    axes[1, 0].set_xlabel("Real Part of x")
    axes[1, 0].set_ylabel("Imaginary Part of x")
    plt.colorbar(im3, ax=axes[1, 0])

    # Plot 4: Imaginary part of the expression
    imag_expr = np.imag(expr_values)
    im4 = axes[1, 1].contourf(real_X, imag_X, imag_expr, levels=50, cmap="coolwarm")
    axes[1, 1].set_title("Imaginary Part of $x^3 - x + 1$")
    axes[1, 1].set_xlabel("Real Part of x")
    axes[1, 1].set_ylabel("Imaginary Part of x")
    plt.colorbar(im4, ax=axes[1, 1])

    plt.tight_layout()
    plt.show()


print("\nVisualizing the complex aspects of the curve...")
visualize_complex_elliptic_curve()
```

```python
def plot_elliptic_curve_as_torus():
    """
    Visualizes the elliptic curve as a torus (donut shape) in the complex domain.
    In complex analysis, elliptic curves are topologically equivalent to a torus.
    This is achieved by considering the complex plane modulo a lattice.
    """
    # The Weierstrass P-function maps C modulo a lattice to the elliptic curve
    # We'll visualize the torus topology that corresponds to the complex structure

    # Parameters for drawing the torus
    R = 2  # Major radius
    r = 1  # Minor radius (R > r for a proper torus)

    # Create parameter grids
    u = np.linspace(0, 2 * np.pi, 100)  # Angle around the torus tube
    v = np.linspace(0, 2 * np.pi, 100)  # Angle around the central hole
    U, V = np.meshgrid(u, v)

    # Parametric equations for a torus
    X = (R + r * np.cos(V)) * np.cos(U)
    Y = (R + r * np.cos(V)) * np.sin(U)
    Z = r * np.sin(V)

    # Create the plot
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection="3d")

    # Plot the torus surface
    ax.plot_surface(
        X,
        Y,
        Z,
        cmap="viridis",
        alpha=0.8,
        rstride=2,
        cstride=2,
        edgecolors="black",
        linewidth=0.1,
    )

    ax.set_title("Elliptic Curve as a Torus (Genus 1 Riemann Surface)", fontsize=16)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    # Set equal aspect ratio
    max_range = (
        np.array([X.max() - X.min(), Y.max() - Y.min(), Z.max() - Z.min()]).max() / 2.0
    )
    mid_x = (X.max() + X.min()) * 0.5
    mid_y = (Y.max() + Y.min()) * 0.5
    mid_z = (Z.max() + Z.min()) * 0.5
    ax.set_xlim(mid_x - max_range, mid_x + max_range)
    ax.set_ylim(mid_y - max_range, mid_y + max_range)
    ax.set_zlim(mid_z - max_range, mid_z + max_range)

    # Set the viewing angle
    ax.view_init(elev=20, azim=30)

    plt.tight_layout()
    plt.show()

    print(
        "\nThe elliptic curve y² = x³ - x + 1, when considered over the complex numbers,"
    )
    print(
        "forms a genus 1 Riemann surface which is topologically equivalent to a torus."
    )
    print("This is due to the Uniformization Theorem and the fact that elliptic curves")
    print("can be represented as C modulo a lattice (period parallelogram).")


print("\nVisualizing the elliptic curve as a torus (the correct complex topology)...")
plot_elliptic_curve_as_torus()
```

```python
def plot_weierstrass_p_function_lattice():
    """
    Illustrate the connection between the complex plane modulo a lattice
    and the elliptic curve via the Weierstrass P-function
    """
    # Define a lattice in the complex plane
    # For this specific curve, we'd normally compute the actual periods
    # but for visualization we'll use a representative lattice
    omega1 = 1.0
    omega2 = 1j

    # Generate points in the fundamental parallelogram
    real_parts = np.linspace(0, 1, 100)
    imag_parts = np.linspace(0, 1, 100)
    real_grid, imag_grid = np.meshgrid(real_parts, imag_parts)

    # Convert to lattice coordinates
    z_real = real_grid * omega1
    z_imag = (
        imag_grid * omega2.imag
    )  # assuming omega2 is purely imaginary for simplicity
    z_lattice = z_real + 1j * z_imag

    # Plot the lattice in the complex plane
    fig = plt.figure(figsize=(12, 5))

    # Plot 1: The lattice in the complex plane
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.scatter(
        z_lattice.real.flatten(),
        z_lattice.imag.flatten(),
        s=1,
        c="blue",
        alpha=0.5,
        label="Lattice points",
    )

    # Draw the fundamental parallelogram
    corners = [0, omega1, omega1 + omega2, omega2, 0]
    ax1.plot(
        [z.real for z in corners],
        [z.imag for z in corners],
        "r-",
        linewidth=2,
        label="Fundamental parallelogram",
    )

    ax1.set_title("Lattice in Complex Plane")
    ax1.set_xlabel("Real Part")
    ax1.set_ylabel("Imaginary Part")
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    # Plot 2: Conceptual mapping to the torus
    ax2: Axes3D = fig.add_subplot(1, 2, 2, projection="3d")

    # Create a torus that represents the identification of opposite edges
    # of the fundamental parallelogram
    R = 2  # Major radius
    r = 0.8  # Minor radius

    # Parameterize the torus
    theta = np.linspace(0, 2.0 * np.pi, 100)
    phi = np.linspace(0, 2.0 * np.pi, 100)
    Theta, Phi = np.meshgrid(theta, phi)

    X_torus = (R + r * np.cos(Theta)) * np.cos(Phi)
    Y_torus = (R + r * np.cos(Theta)) * np.sin(Phi)
    Z_torus = r * np.sin(Theta)

    ax2.plot_surface(X_torus, Y_torus, Z_torus, alpha=0.7, cmap="viridis")
    ax2.set_title(
        "Torus from Period Lattice\n(C ≅ Parallelogram with Opposite Edges Identified)"
    )

    plt.tight_layout()
    plt.show()

    print("\nMathematical Explanation:")
    print(
        "An elliptic curve over ℂ is ℂ/Λ where Λ is a lattice (discrete subgroup of ℂ)."
    )
    print(
        "The fundamental parallelogram formed by ω₁ and ω₂ is mapped to a torus by identifying opposite edges."
    )
    print(
        "The Weierstrass ℘-function provides the bijection between ℂ/Λ and the elliptic curve."
    )


print(
    "\nShowing the relationship between the complex plane modulo a lattice and the torus..."
)
plot_weierstrass_p_function_lattice()
```

```python
def j_invariant_of_curve():
    """
    Calculate the j-invariant of the elliptic curve y² = x³ - x + 1
    """
    # For curve y² = x³ + ax + b, we have a = -1, b = 1
    a = -1
    b = 1

    # Calculate discriminant and j-invariant
    discriminant = -16 * (4 * a**3 + 27 * b**2)  # Simplified formula for this form
    c4 = -27 * (4 * a**3 + 27 * b**2)  # Needed for j-invariant
    j_inv = c4 / discriminant * 1728  # j-invariant

    print("\nj-invariant of y² = x³ - x + 1:")
    print(f"a = {a}, b = {b}")
    print(f"Discriminant = {discriminant}")
    print(f"j-invariant = {j_inv}")


j_invariant_of_curve()
```
